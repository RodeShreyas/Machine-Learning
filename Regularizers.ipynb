```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BostonHousing.csv')
```

```python
df
```

```python
X = df.iloc[:,0:13].values
```

```python
X
```

# Checking Nan Values in X

```python
nan_mask = np.isnan(X)
print(f'Number of NaNs in X:{np.sum(nan_mask)}')
```

```python
y = df.iloc[:,13].values
```

```python
y
```

```python
X_cleaned = X[~np.isnan(X).any(axis=1)]
Y_cleaned = y[~np.isnan(X).any(axis=1)]
```

## Filling missing values by calculating mean

```python
col_mean = np.nanmean(X, axis=0)
indices = np.where(np.isnan(X))
X_filled = X.copy()
X_filled[indices] = np.take(col_mean, indices[1])
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
mse = cross_val_score(lin_reg,X_cleaned,Y_cleaned,scoring='neg_mean_squared_error',cv=5)
```

```python
mse
```

```python
np.mean(mse)
```

# Ridge

```python
from sklearn.linear_model import Ridge
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
```

```python
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
```

```python
ridge_regressor = GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
```

```python
ridge_regressor.fit(X_cleaned,Y_cleaned)
```

```python
ridge_regressor.best_params_
```

```python
ridge_regressor.best_score_
```

# Lasso

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso()
```

```python
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
```
