# Ploynomial Regression

## polynomial regression is a form of regression analysis in which the relationship between the independent variable x and the dependent variable y is modeled as an nth degree polynomial in x

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
sns.set_style('darkgrid')
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:,1].values
```

```python
x 
```

```python
y = dataset.iloc[:,2].values
```

```python
y 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x.reshape(-1,1),y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x.reshape(-1,1)),color='green')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('LINEAR REGRESSION')
plt.show()
```

```python
x 
```

```python
x.reshape(-1,1)
```

```python
lin_reg.predict(np.array([7]).reshape(-1,1))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=4)   # hyperparameter = 1,2,3,4,5,6,7,8,9
```

```python
X_poly = poly_reg.fit_transform(x.reshape(-1,1))
```

```python
X_poly
```

```python
lin_reg2 = LinearRegression()
```

```python
lin_reg2.fit(X_poly,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(poly_reg.fit_transform(x.reshape(-1,1))),color='green')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('POLYNOMIAL REGRESSION')
plt.show()
```

```python
pred = 6 
pred2 = np.array([[pred]])
lin_reg2.predict(poly_reg.fit_transform(pred2))
```
