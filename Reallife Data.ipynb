# Categorical variables

```python
data_no_Multicolinearity.describe(include='all')
```

```python
data_with_dummies = pd.get_dummies(data_no_Multicolinearity,dtype='int',drop_first=True)
```

```python
data_with_dummies
```

```python
target = data_with_dummies['log_price']    # y = dep
```

```python
target
```

```python
inputs = data_with_dummies.drop(['log_price'],axis=1)
```

```python
inputs
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(inputs)
```

```python
input_scaled = scaler.transform(inputs)
```

```python
input_scaled
```

```python
df = pd.DataFrame(input_scaled)
```

```python
df 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(input_scaled,target,test_size=0.2,random_state=365)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
yhat = reg.predict(X_train)
```

```python
plt.scatter(y_train,yhat)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.show()
```

```python
sns.distplot(y_train-yhat)
```

```python
reg.score(X_train,y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```
